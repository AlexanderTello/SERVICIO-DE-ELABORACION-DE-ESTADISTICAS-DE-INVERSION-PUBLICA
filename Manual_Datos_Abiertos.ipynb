# **Datos abiertos**
Los “Datos Abiertos” consisten en la información gubernamental disponible al público a través de internet en formato digital, estandarizado y comparable. 
Para acceder a los datos abiertos no es necesario desarrollar ninguna herramienta especial como en los casos anteriores. Esto ya que existen robustas librerías escritas en lenguaje Python, como **requests**, enfocadas a la extracción de estos archivos. 
A continuación, se presentan las formas de lectura de archivos en formato Excel y CSV haciendo uso de las librerías **urllib y requests**.

### **1. Importar las librerias**

La librería a importar es **urllib**, la cual ya se encuentra por default en el entorno, pero para evitar inconvenientes con la versión de Python que se use se debe ordenar su importación. Esta librería tiene métodos como **requests y urlopen** que permiten obtener la respuesta de una búsqueda web.  
Asimismo, se importa la librería pandas que servirá para leer los archivos que los paquetes extraígan. 

```python
from urllib.request import Request, urlopen
import requests #Libreria para hacer consultas online
import pandas as pd #Libreria para manipular bases de datos
```

### **2. Lectura web de archivos en formato Excel**

La principal ventaja de este procedimiento es que se evita el desgaste de recursos de la computadora en la descarga del archivo, y se ahorra el tiempo en que despues de su descarga debería iportarse al entorno.  
Para este procedimiento se utiliza la librería "requests". El único insumo necesario es el vínculo donde se enceuntra ubicado el archivo. 
En este caso, por ejemplo se extrae una base de datos abiertos del CONOSCE. 
<center><img src="https://raw.githubusercontent.com/AlexanderTello/imagenes/main/6.PNG" width="640" align="center" alt="1" /></center>

```python
url = 'https://tinyurl.com/conoscepac2018' #Enlace web del archivo 
consulta = requests.get(url)               #Vincualción de Python con la web mediante la librería request
open('CONOSCE.xlsx', 'wb').write(consulta.content) #Extracción del archivo Excel 
df = pd.read_excel('CONOSCE.xlsx',header=0, skiprows=[0] ) #Lectura del archivo
df.head()                                  #Visualización de las 5 primeras filas
```

### **3. Lectura web de archivos en formato CSV**

Al igual que en el caso anterior, con este procedimiento se evita el desgaste de recursos de la computadora en la descarga del archivo, y se ahorra el tiempo en que despues de su descarga debería iportarse al entorno.  
Para este procedimiento se utiliza la librería "urllib y sus dependencias como urlopen". El único insumo necesario es el vínculo donde se enceuntra ubicado el archivo. 
En este caso, por ejemplo se extrae una base de datos abiertos de Inversiones. 
<center><img src="https://raw.githubusercontent.com/AlexanderTello/imagenes/main/7.PNG" width="640" align="center" alt="1" /></center>

```python
link = "https://storage.googleapis.com/mef-peru/InversionPublica/DETALLE_INVERSIONES.csv" #Enlace web del archivo 
busqueda = Request(link)                        #Vincualción de Python con la web mediante la librería request
busqueda.add_header('User-Agent','Mozilla/5.0') #Se especifica desde qué navegador se desea realizar la búsqueda
contenido = urlopen(busqueda)                          #Extracción del archivo 
inversiones = pd.read_csv(contenido, sep = ',',encoding='latin-1')#Lectura del archivo
inversiones.head()                              #Visualización de las 5 primeras fila
```
